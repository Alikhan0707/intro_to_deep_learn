#### Используем набор примеров fashion-MNIST

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

from keras.datasets import mnist

from keras.layers import Dense
from keras.models import Sequential

from keras.utils import to_categorical
```

```python
(X_train, y_train_label), (X_test, y_test_label) = mnist.load_data()
```

```python
X_train_max = X_train.max()
X_test_max = X_test.max()
X_train_max, X_test_max
```

###### Нормировка

```python
X_train = X_train / X_train_max
X_test = X_test / X_test_max
```

```python
X_test.max()
```

```python
shape_of_X_train = X_train.shape
new_shape = shape_of_X_train[-1] * shape_of_X_train[-2]
```

###### Вытягивание

```python
X_train = X_train.reshape((-1, new_shape))
X_test = X_test.reshape((-1, new_shape))
```

```python
y_train = to_categorical(y_train_label)
y_test = to_categorical(y_test_label)
```

```python
model = Sequential()

model.add(Dense(64, input_shape=(new_shape, ), activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(10, activation='sigmoid'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='sgd')
```

```python
history = model.fit(X_train, y_train, 
                    epochs=5,
                    batch_size=10,
                    verbose=1, validation_split=0.2
                   )
```

#### 2.  Проверьте работу разных оптимизаторов (SGD, Adam, RMSProp) для одной из моделей п.1. Фиксируйте для тренировочного и тестового набора метрики accuracy.

```python
from keras.optimizers import SGD, Adagrad, RMSprop, Adam

model = Sequential()

model.add(Dense(64, input_shape=(new_shape, ), activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(10, activation='sigmoid'))
model.compile(loss='categorical_crossentropy', 
              optimizer='sgd', 
              metrics=['accuracy'])

model.save_weights('model_weights.h5')

optimizers_list = [SGD, Adagrad, Adam, RMSprop]

plt.figure(figsize=(16,7))

models = {}

epoch_nums = 5
batch_size = 10

epoch = np.arange(epoch_nums + 1)

for optimizer in optimizers_list:
    
    model = Sequential()

    model.add(Dense(64, activation='relu', input_shape=(new_shape, )))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(10, activation='sigmoid'))
    
    model.load_weights('model_weights.h5')
    
    model.compile(
        optimizer=optimizer(), 
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    
    zero_loss_train = model.evaluate(X_train, y_train, verbose=0)
    zero_loss_valid = model.evaluate(X_test, y_test, verbose=0)

    history = model.fit(X_train, y_train, 
                        epochs=epoch_nums,
                        batch_size=batch_size,
                        validation_data=(X_test, y_test),
                        verbose=1)
    
    train_loss = [zero_loss_train[0]] + history.history['loss']
    valid_loss = [zero_loss_valid[0]] + history.history['val_loss']
    
    train_loss = np.log(train_loss)
    valid_loss = np.log(valid_loss)
    
    optimizer_name = model.optimizer.get_config()['name']
    
    models[optimizer_name] = model
    
    plt.plot(epoch, train_loss, '-', label=optimizer_name + 'train')
    plt.plot(epoch, valid_loss, '--', label=optimizer_name + 'valid')

plt.legend()
plt.title('нейросети от оптимизатора log(loss)')
plt.xlabel('epoch')
plt.ylabel('log(loss)')
plt.show()
```
